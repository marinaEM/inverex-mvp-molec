## Interpretation Notes

- The LightGBM model is trained on breast cell-line inhibition labels, not patient treatment response.
- Even when gene features dominate aggregate importance, the model still expresses a global potency prior that recurs across patients.
- Mutation/CNV, subtype, and clinical actionability are absent from the training matrix and therefore must be layered in explicitly at ranking time.
- For this reason the project now uses LightGBM only as one component inside a composite personalized ranker.
